```python
import pandas as pd
import numpy as np


df = pd.read_csv('bse_data.csv')
print(df.head(5))
```

```python
print(df.isnull().sum())
```

```python
# FILL THE MISSING CELLS IN LIVING ALLOWANCE COLUMN WITH THE MODE VALUE IN THE COLUMN

df['living_allowance'] = df['living_allowance'].fillna(df['living_allowance'].mode())
print(df.head(5))
```

```python
#CHECK FOR DUPLICATES

print(df.duplicated().sum())
```

```python
print(df.describe())
```

```python
#  COUNT NUMBER OF THOSE WITH AND THOSE WITHOUT BOYFRIENDS AND GIRLFRIENDS

print(df['have_bf_gf'].value_counts())
```

```python
# Count Males and females in the dataset
print(df['gender'].value_counts())
```

```python
print(df['daily_meals'].value_counts())
```

```python
# have bf gf against gender
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x='gender',y='have_bf_gf', data=df)
plt.title('Having bf/gf rate by gender')
plt.show()
```

```python
# How often do people with girlfriends and boyfriends eat

sns.barplot(x='daily_meals',y='have_bf_gf', data=df)
plt.title('Have a bf/gf?, how often do you eat')
plt.show()
```

```python
# Gender distribution

sns.histplot(df['gender'],bins=30, kde=True)
plt.title('Gender distribution')
plt.show()
```

```python
siblings = df['number_siblings'].value_counts()

# create pie chart
plt.figure(figsize=(8,8))
plt.pie(siblings, labels=siblings.index, autopct='%1.1f%%', startangle=90)
plt.title('Number of siblings chart')
plt. axis('equal')
plt.show()
```

```python
# PIE CHART FOR AGE DISTRIBUTION

age_count = df['age'].value_counts()

plt.figure(figsize=(8,8))
plt.pie(age_count, labels=age_count.index, autopct='%1.1f%%', startangle=90)
plt.title('Age distribution')
plt.axis('equal')
plt.show()
           
```
