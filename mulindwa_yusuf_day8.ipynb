```python
# Data science with python
'''
 Data science is an interdisciplinary field that combines various techniques
 from statistics , computer science and domain-specific knowledge to extract
 meaningful insights from data.
 It involves the use of algorithms,data analysis,machine learning and other  advanced methods to analyze and interpret complex
 data sets
'''
# Application of data science
# step 1    
# importing Necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Step 2
# loading the data
df = pd.read_csv('crop.csv')

print(df.head(5))

```

```python
# Step 3: Data cleaning and pre-processing

# Check for missing values

print(df.isnull().sum())
```

```python
# Check for missing values using heatmap

missing = df.isnull()

plt.figure(figsize=(8,8))
sns.heatmap(missing, cmap='viridis', cbar=False)
plt.title('Missing values heatmap')
plt.show()
           
```

```python
# checking for duplicates

print(df.duplicated().sum())
```

```python
print(df.describe())
```

```python
# Plotting temperature distribution

sns.histplot(df['temperature'],color='yellow',bins=30, kde=True)
plt.title('Temperature Distribution')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('crop.csv')

c=df.label.astype('category')
targets = dict(enumerate(c.cat.categories))
df['target'] = c.cat.codes
y= df['target']
x = df[['N','P','K','temperature','humidity','ph','rainfall']]

# PLOT CORELATION

sns.heatmap(x.corr(), annot=True , cmap='coolwarm')
plt.title('Corelation')
plt.show()
```
